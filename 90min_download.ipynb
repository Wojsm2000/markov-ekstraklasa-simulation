```python
def return_match_data():
    page = requests.get("http://www.90minut.pl/liga/1/liga14072.html")
    soup = BeautifulSoup(page.content, features="html.parser")
    tb=soup.find_all("table", {"class": "main"})
    home_teams=[]
    away_teams=[]
    scores=[]
    dates=[]
    for i in tb:
        d=i.find_all("td", {"valign": "top"})
        if len(d)>0:
            for i in range(36):
                if i%4==0:
                    home_teams.append(d[i].text.strip())
                elif i%4==1:
                    scores.append(d[i].text.strip())
                elif i%4==2:
                    away_teams.append(d[i].text.strip())
                else:
                    dates.append(d[i].text.strip())
    home_score=[]
    away_score=[]
    for i in scores:
        if i!="-":
            home_score.append(int(i.split("-")[0]))
            away_score.append(int(i.split("-")[1]))
    else:
        home_score.append(-1)
        away_score.append(-1)
    return {"home_teams": home_teams, "away_teams": away_teams, "home_scores": home_score, "away_scores": away_score, "dates": dates}
```

```python
json.loads(json.dumps(return_match_data(), indent=4))
```

```python
def return_table():
    page = requests.get("http://www.90minut.pl/liga/1/liga14072.html")
    soup = BeautifulSoup(page.content, features="html.parser")
    tb=[]
    teams = []
    points = []
    tb=soup.find_all("table", {"class": "main2"})[0]
    tb=tb.find_all("tr",{"align": "center"})
    for i in tb:
        teams.append(i.find_all("td")[1].text.lstrip('\xa0').strip())
        points.append(i.find_all("td")[3].text)
    tb=dict(zip(teams[2:], points[2:]))
    return tb
        
```
