```python
import requests
import json
from io import StringIO
import pandas as pd
```

```python
URL = "http://api.clubelo.com/2026-02-11"
```

```python
r = requests.get(url = URL)
```

```python

if r.status_code == 200:
    data = StringIO(r.text)
data=pd.read_csv(data)
```

```python
data = data[data["Country"] == "POL"][["Club", "Elo"]]

with open('elo_data.json', 'w', encoding='utf-8') as f:
    json.dump(data.to_dict(orient="records"), f, indent=4, ensure_ascii=False)
```

```python
def load_elo_data(date):
    URL = f"http://api.clubelo.com/{date}"
    r = requests.get(url = URL)
    if r.status_code == 200:
        data = StringIO(r.text)
    data=pd.read_csv(data)
    data = data[data["Country"] == "POL"][["Club", "Elo"]]
    return data.to_dict(orient="records")
```

```python
load_elo_data("2026-02-11")
```
